## 1. Why Python for Economists?

### The shift in economics research

For decades economists used Stata for microeconometrics, MATLAB for macroeconomic models, and Excel for everything else. Over the last ten years Python has become the dominant tool for computational economics, for several reasons:

| Tool | Strength | Limitation |
|---|---|---|
| Stata | Panel data, microeconometrics | Expensive license, limited visualization |
| MATLAB | Numerical methods, fast matrices | Very expensive, less community |
| R | Statistics, econometrics | Slower for large simulations |
| Excel | Accessible, ubiquitous | Cannot handle large data or simulations |
| **Python** | **Free, general-purpose, vast ecosystem** | **Steeper initial learning curve** |

### Who uses Python in economics?

- **Central banks**: Bank of England, Federal Reserve, ECB — all maintain Python codebases for macro models and stress testing
- **Academia**: The [QuantEcon project](https://quantecon.org) by Nobel laureate Thomas Sargent and John Stachurski provides free Python textbooks for graduate macroeconomics
- **International organisations**: IMF, World Bank, OECD publish Python tools for data analysis
- **Finance industry**: Portfolio management, risk, algorithmic trading

### What Python is NOT the best choice for

- Production databases and SQL-heavy work (use SQL directly)
- Very large-scale statistical consulting where R packages dominate
- Quick one-off tables for a paper (Excel is fine for that)

Python is best when you need to **combine data wrangling + simulation + estimation + visualisation** in one reproducible workflow — exactly what computational economics requires.

## 5. The Sugarscape Running Thread

Throughout this series we use the **Sugarscape G1MT model** (Epstein & Axtell, 1996) as a connecting thread. Sugarscape is a classic agent-based model in which artificial traders move across a landscape, harvest resources (sugar and spice), and trade with one another.

Why Sugarscape?
- It demonstrates **emergent inequality**: wealth concentrations arise from simple trading rules, not from any top-down design
- It uses nearly every tool in this series simultaneously: NumPy arrays, Pandas DataFrames, Matplotlib figures, Mesa agents, NetworkX trade networks
- It is implemented step-by-step in the companion **Session_3 through Session_20** notebooks in this repository

**You do not need to understand Sugarscape to start.** Each notebook stands alone. References to Sugarscape appear as optional links showing how the concepts you are learning appear in a real research model.

**Reference:** Epstein, J. M., & Axtell, R. (1996). *Growing Artificial Societies: Social Science from the Bottom Up.* MIT Press.

## 6. Code Style in This Series

Every notebook in this series follows the same conventions. Knowing them will help you read any cell quickly.

### Explicit imports — never `import *`
```python
# Good — you can always see where a function comes from
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Bad — hides where functions come from, causes name conflicts
from numpy import *
```

### Economics variable names
We use full economic vocabulary in variable names so the code reads like an economics paper:
```python
# Good
gdp = consumption + investment + government_spending + net_exports

# Bad — meaningless outside this context
y = a + b + c + d
```

### Structure of every section
1. **Markdown explanation** — what the economic concept is and what the code will do
2. **Worked example** — complete, runnable code with comments
3. **Your Turn** — an exercise for you to attempt
4. **Solution** — a collapsed cell you can expand to check your answer

### Code grows gradually
Early cells in each notebook are 3–8 lines. Later cells grow to 20–35 lines. We never introduce complexity without first showing the simpler version.

### Package installation
Each notebook installs its required packages at the top using the `try/except` pattern:
```python
try:
    import numpy as np
except ImportError:
    !pip install numpy --quiet
    import numpy as np
```
This means the notebook works whether or not the package is already installed.

## 7. Quick Python Survival Cheatsheet

### Getting help
```python
?len          # one-line description of the len function
??len         # full source code of len (if available)
help(len)     # verbose documentation
# In a cell, type the function name and press Shift+Tab for a popup
```

### Common error messages and what they mean

| Error | Meaning | Likely cause |
|---|---|---|
| `NameError: name 'x' is not defined` | You used a variable before creating it | Run the earlier cells first, or check spelling |
| `TypeError: unsupported operand type` | You mixed incompatible types (e.g., string + number) | Check your variable types with `type(x)` |
| `IndentationError` | Python is strict about indentation | Use 4 spaces, never tabs |
| `ModuleNotFoundError` | Package not installed | Run the `!pip install` cell at the top |
| `IndexError: list index out of range` | You tried to access position 5 in a 5-element list | Remember: Python counts from 0, so the last element is index 4 |
| `KeyError` | Dictionary key does not exist | Print `dict.keys()` to see what keys are available |

### How to restart if something goes wrong
- **Runtime → Restart runtime** — clears all variables and imported packages
- After restarting, run all cells from the top: **Runtime → Run all**

### The most important Python rules for economists
1. Python **counts from zero**: the first element of a list `data` is `data[0]`, not `data[1]`
2. Python **indentation is syntax**: blocks of code (loops, functions, if statements) are marked by indentation, not by `do`/`end` keywords
3. **Assignment** uses `=`; **comparison** uses `==`
4. Comments start with `#` — Python ignores everything after `#` on a line

### A one-minute demo: Python computing GDP

Before you move to NB_01, here is a taste of what Python code looks like. Run this cell — no prior knowledge needed, just observe.

````python
# Expenditure approach to GDP (billions USD, approximate 2023 values)
consumption         = 17_800   # C: household consumption
investment          =  4_200   # I: private investment
government_spending =  4_600   # G: government expenditure
net_exports         =   -800   # NX: exports minus imports

gdp = consumption + investment + government_spending + net_exports

print(f"GDP (expenditure approach): ${gdp:,} billion")
print(f"Consumption share of GDP:   {consumption/gdp:.1%}")
print(f"Investment share of GDP:    {investment/gdp:.1%}")
print(f"Government share of GDP:    {government_spending/gdp:.1%}")
````
