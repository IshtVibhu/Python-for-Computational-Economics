## 1. Creating Arrays

A NumPy **array** is a grid of values, all of the same type. Unlike Python lists, arrays support fast vectorised operations.

````python
# 1-D array from a list
gdp_growth = np.array([2.1, 2.4, -3.4, 5.7, 2.9, 1.8])  # annual % growth
print(f"Array:  {gdp_growth}")
print(f"Shape:  {gdp_growth.shape}")   # (6,) — one dimension, 6 elements
print(f"Dtype:  {gdp_growth.dtype}")   # float64

# Useful constructors
zeros   = np.zeros(5)              # five zeros
ones    = np.ones((3, 4))          # 3x4 matrix of ones
seq     = np.arange(0, 1.1, 0.1)  # 0.0, 0.1, 0.2, ..., 1.0
linseq  = np.linspace(0, 1, 11)   # same, but specifying number of points

print(f"\narange(0, 1.1, 0.1): {seq.round(1)}")
print(f"linspace(0, 1, 11):  {linseq.round(1)}")
````

## 2. Vectorised Arithmetic

With NumPy, arithmetic on arrays applies **element-wise** — no loop required. This is both faster and more readable.

````python
# Prices and quantities for 5 goods
prices    = np.array([2.0, 5.0, 1.5, 8.0, 3.0])
quantities = np.array([100, 40, 200, 20, 80])

# Element-wise — each price times its quantity
revenues = prices * quantities
print(f"Revenues per good: {revenues}")
print(f"Total revenue:     {revenues.sum():.0f}")

# Comparison — returns a boolean array
high_value = revenues > 200
print(f"\nHigh-value goods mask: {high_value}")
print(f"High-value revenues:   {revenues[high_value]}")

# Broadcasting: add 10% tax to all prices
tax_rate = 0.10
after_tax_prices = prices * (1 + tax_rate)
print(f"\nPrices after 10% tax: {after_tax_prices}")
````

## 3. 2-D Arrays and the Sugarscape Landscape

The Sugarscape world is a 50×50 grid. Each cell holds a sugar capacity and a spice capacity. This is a 2-D NumPy array.

````python
import numpy as np

GRID_SIZE = 50

# Sugarscape-style landscape: sugar peaks in top-left and bottom-right
def make_sugar_landscape(size=50):
    """Create a sugar landscape with two peaks, as in Epstein & Axtell (1996)."""
    landscape = np.zeros((size, size))
    peak1 = (15, 15)   # (row, col)
    peak2 = (35, 35)
    for r in range(size):
        for c in range(size):
            d1 = np.sqrt((r - peak1[0])**2 + (c - peak1[1])**2)
            d2 = np.sqrt((r - peak2[0])**2 + (c - peak2[1])**2)
            landscape[r, c] = max(4 - d1 / 5, 4 - d2 / 5, 0)
    return np.clip(landscape, 0, 4).round().astype(int)

sugar = make_sugar_landscape(GRID_SIZE)

print(f"Landscape shape: {sugar.shape}")
print(f"Sugar capacity range: {sugar.min()} to {sugar.max()}")
print(f"Mean sugar:       {sugar.mean():.2f}")
print(f"Total sugar:      {sugar.sum()}")
print(f"\nTop-left 8x8 corner:")
print(sugar[:8, :8])
````

## 4. Linear Algebra — Solving Economic Systems

Many economic models reduce to solving a linear system `Ax = b`. NumPy solves these in one line.

````python
# Supply and demand equilibrium for two related goods
# Demand system (own-price and cross-price elasticities embedded):
#   Q1 = 100 - 3*P1 + 1*P2
#   Q2 = 120 + 1*P1 - 4*P2
# Supply system:
#   Q1 =  2*P1
#   Q2 =  3*P2
# Setting demand = supply and rearranging to Ax = b:
#   5*P1 - 1*P2 = 100
#  -1*P1 + 7*P2 = 120

A = np.array([[5, -1],
              [-1,  7]], dtype=float)
b = np.array([100, 120], dtype=float)

# Solve the system
prices = np.linalg.solve(A, b)
P1, P2 = prices

Q1_star = 2 * P1
Q2_star = 3 * P2

print("General Equilibrium Prices and Quantities:")
print(f"  P1* = {P1:.2f},  Q1* = {Q1_star:.2f}")
print(f"  P2* = {P2:.2f},  Q2* = {Q2_star:.2f}")

# Verify: residuals should be near zero
residual = A @ prices - b
print(f"\nResidual (should be ~0): {residual}")
````

## 5. Monte Carlo Simulation — Economic Uncertainty

Monte Carlo methods simulate thousands of possible futures. Here we estimate the distribution of GDP after 10 years of uncertain growth.

````python
np.random.seed(2024)

N_SIMULATIONS = 100_000
N_YEARS       = 10
INITIAL_GDP   = 1000.0
MEAN_GROWTH   = 0.025   # 2.5% per year
STD_GROWTH    = 0.020   # 2.0% standard deviation

# Draw N_SIMULATIONS * N_YEARS random growth rates at once — vectorised
annual_growth = np.random.normal(MEAN_GROWTH, STD_GROWTH,
                                  size=(N_SIMULATIONS, N_YEARS))

# Compound growth: multiply (1 + rate) across years for each simulation
gdp_final = INITIAL_GDP * np.prod(1 + annual_growth, axis=1)

print("Monte Carlo GDP simulation (10 years, 100,000 paths):")
print(f"  Mean final GDP:   {gdp_final.mean():.1f}")
print(f"  Median:           {np.median(gdp_final):.1f}")
print(f"  Std deviation:    {gdp_final.std():.1f}")
print(f"  5th percentile:   {np.percentile(gdp_final, 5):.1f}")
print(f"  95th percentile:  {np.percentile(gdp_final, 95):.1f}")
prob_contraction = (gdp_final < INITIAL_GDP).mean()
print(f"  P(GDP < initial): {prob_contraction:.1%}")
````

## 6. Your Turn — Wealth Distribution Simulation

Simulate 1,000 agents each starting with 50 units of wealth. Each step, every agent's wealth grows by a **random multiplicative factor** drawn from a log-normal distribution with mean 1.0 and sigma 0.2. Run 20 steps and report:
- Final mean wealth
- Gini coefficient of the final wealth distribution

**Hint for the Gini coefficient:** For a sorted array `w`, the Gini coefficient is:
```
G = (2 * sum(i * w[i]) / (n * sum(w))) - (n+1)/n
```
where `i` is the 1-based rank.

````python
# Your code here
````

````python
#@title Solution
np.random.seed(42)
N_AGENTS = 1000
N_STEPS  = 20
wealth   = np.full(N_AGENTS, 50.0)

for _ in range(N_STEPS):
    growth_factors = np.random.lognormal(mean=0.0, sigma=0.2, size=N_AGENTS)
    wealth *= growth_factors

def gini(w):
    w_sorted = np.sort(w)
    n = len(w_sorted)
    ranks = np.arange(1, n + 1)
    return (2 * np.sum(ranks * w_sorted) / (n * w_sorted.sum())) - (n + 1) / n

print(f"Final mean wealth:    {wealth.mean():.1f}")
print(f"Final median wealth:  {np.median(wealth):.1f}")
print(f"Gini coefficient:     {gini(wealth):.3f}")
print("\nWealth skews right: multiplicative growth generates inequality.")
````
