## 1. Variables and Arithmetic

In Python, a **variable** is a name that stores a value. You create one with `=`. Python infers the type automatically — you never need to declare it.

```python
# Basic economic variables
price = 10.50          # float (decimal number)
quantity = 200         # int (whole number)
good_name = "wheat"    # str (text)
is_normal_good = True  # bool (True or False)

total_revenue = price * quantity
print(f"Total revenue from {good_name}: ${total_revenue:,.2f}")
print(f"Is it a normal good? {is_normal_good}")
print(f"Variable types: price={type(price).__name__}, quantity={type(quantity).__name__}")
```

### Arithmetic operators

| Operator | Symbol | Example | Result |
|---|---|---|---|
| Addition | `+` | `3 + 2` | `5` |
| Subtraction | `-` | `10 - 4` | `6` |
| Multiplication | `*` | `3 * 4` | `12` |
| Division | `/` | `7 / 2` | `3.5` |
| Integer division | `//` | `7 // 2` | `3` |
| Remainder (modulo) | `%` | `7 % 2` | `1` |
| Exponentiation | `**` | `2 ** 10` | `1024` |

```python
# Cobb-Douglas production function: Y = A * K^alpha * L^(1-alpha)
A = 1.0      # total factor productivity
K = 100.0    # capital
L = 50.0     # labour
alpha = 0.33 # capital share

Y = A * K**alpha * L**(1 - alpha)
print(f"Output Y = {Y:.2f}")
print(f"Capital share of output: {alpha:.0%}")
print(f"Labour share of output:  {1-alpha:.0%}")
```

## 2. Lists, Tuples, and Dictionaries

These three containers are the workhorses of Python data handling before you reach NumPy and Pandas.

```python
# List — ordered, mutable (you can change it after creation)
gdp_growth_rates = [2.1, 2.4, -3.4, 5.7, 2.9]  # % per year

print(f"First year growth:  {gdp_growth_rates[0]}%")   # index 0 = first element
print(f"Last year growth:   {gdp_growth_rates[-1]}%")  # index -1 = last element
print(f"Number of years:    {len(gdp_growth_rates)}")
print(f"Average growth:     {sum(gdp_growth_rates)/len(gdp_growth_rates):.2f}%")

# Slicing: list[start:stop] — stop is exclusive
post_recession = gdp_growth_rates[3:]  # from index 3 onwards
print(f"Post-recession rates: {post_recession}")

# Adding elements
gdp_growth_rates.append(1.8)
print(f"After appending new year: {gdp_growth_rates}")
```

```python
# Dictionary — maps keys to values; ideal for named economic data
country_gdp = {
    "United States": 26_854,
    "China":         17_701,
    "Germany":        4_430,
    "Japan":          4_231,
    "India":          3_733,
}  # GDP in billion USD, 2023

print(f"US GDP: ${country_gdp['United States']:,} billion")
print(f"Countries tracked: {list(country_gdp.keys())}")

# Add a new entry
country_gdp["United Kingdom"] = 3_089
print(f"\nLargest economy: {max(country_gdp, key=country_gdp.get)}")

# Iterate over a dictionary
world_total = sum(country_gdp.values())
print(f"\nGDP shares (of tracked countries):")
for country, gdp in sorted(country_gdp.items(), key=lambda x: -x[1]):
    print(f"  {country:<20} {gdp/world_total:.1%}")
```

## 3. Conditional Logic

Conditional statements let your code make decisions based on values.

```python
# Classify business cycle phase based on GDP growth
gdp_growth = -1.2  # percent

if gdp_growth >= 3.0:
    phase = "Boom"
elif gdp_growth >= 0.5:
    phase = "Expansion"
elif gdp_growth >= -0.5:
    phase = "Slowdown"
else:
    phase = "Recession"

print(f"GDP growth {gdp_growth:.1f}% → Business cycle phase: {phase}")
```

```python
# Comparison operators
x = 10
print(x > 5)    # True
print(x == 10)  # True  (== is comparison, = is assignment)
print(x != 7)   # True  (!= means 'not equal')

# Logical operators: and, or, not
inflation = 4.2
unemployment = 4.8

if inflation > 3 and unemployment < 5:
    print("Potential stagflation signal: raise rates")
elif inflation <= 2 or unemployment > 6:
    print("Consider easing")
else:
    print("Economy near dual mandate target")
```

## 4. Loops

Loops repeat a block of code. They are fundamental to simulation — every time step in a model is a loop iteration.

```python
# Simulate compound growth with a for loop
gdp = 1000.0      # starting GDP (arbitrary units)
growth_rate = 0.025  # 2.5% per year
years = 10

print("Year | GDP")
print("-" * 20)
for year in range(years + 1):   # range(11) gives 0, 1, 2, ..., 10
    print(f"{year:4d} | {gdp:>10.2f}")
    gdp = gdp * (1 + growth_rate)
```

```python
# List comprehension — a compact for-loop that builds a list
# Compute doubling times for a range of growth rates
growth_rates = [1, 2, 3, 4, 5, 6, 7]
doubling_times = [70 / r for r in growth_rates]

for rate, dt in zip(growth_rates, doubling_times):
    print(f"{rate}% growth → doubles in {dt:.1f} years")
```

```python
# While loop — repeat until a condition is met
# How many years until GDP triples?
gdp = 1.0
target = 3.0
growth_rate = 0.025
years_elapsed = 0

while gdp < target:
    gdp *= (1 + growth_rate)
    years_elapsed += 1

print(f"GDP triples in {years_elapsed} years at {growth_rate:.1%} annual growth.")
```

## 5. Functions

A **function** packages code so you can reuse it. Well-named functions make economic models readable.

```python
def cobb_douglas(K, L, A=1.0, alpha=0.33):
    """Cobb-Douglas production function Y = A * K^alpha * L^(1-alpha)."""
    return A * K**alpha * L**(1 - alpha)

def marginal_product_capital(K, L, A=1.0, alpha=0.33):
    """MPK = alpha * Y / K"""
    Y = cobb_douglas(K, L, A, alpha)
    return alpha * Y / K

def marginal_product_labour(K, L, A=1.0, alpha=0.33):
    """MPL = (1 - alpha) * Y / L"""
    Y = cobb_douglas(K, L, A, alpha)
    return (1 - alpha) * Y / L

# Try different capital/labour ratios
for K in [50, 100, 200]:
    Y = cobb_douglas(K, L=100)
    mpk = marginal_product_capital(K, L=100)
    print(f"K={K:3d}  Y={Y:.1f}  MPK={mpk:.4f}")

print("\nDiminishing returns to capital: MPK falls as K rises.")
```

## 6. Classes — Building Your First Economic Agent

A **class** is a blueprint for creating objects. In agent-based modelling, every agent is an instance of a class. This section builds a minimal `Trader` — the foundation of the Sugarscape model.

```python
import random

class Trader:
    """A minimal economic agent who holds and trades two goods."""

    def __init__(self, trader_id, sugar, spice, metabolism_sugar, metabolism_spice):
        self.trader_id = trader_id
        self.sugar = sugar
        self.spice = spice
        self.metabolism_sugar = metabolism_sugar  # units consumed per step
        self.metabolism_spice = metabolism_spice
        self.alive = True

    @property
    def wealth(self):
        """Total wealth = sum of both endowments."""
        return self.sugar + self.spice

    @property
    def mrs(self):
        """Marginal rate of substitution: how many units of spice the agent
        would give up for one unit of sugar (Epstein-Axtell rule)."""
        if self.spice == 0:
            return float('inf')
        return (self.metabolism_sugar / self.metabolism_spice) * (self.spice / self.sugar) if self.sugar > 0 else float('inf')

    def consume(self):
        """Each step the trader burns their metabolic needs."""
        self.sugar -= self.metabolism_sugar
        self.spice -= self.metabolism_spice
        if self.sugar <= 0 or self.spice <= 0:
            self.alive = False

    def __repr__(self):
        return (f"Trader({self.trader_id}) "
                f"sugar={self.sugar:.1f} spice={self.spice:.1f} "
                f"wealth={self.wealth:.1f} alive={self.alive}")


# Create a small population
random.seed(42)
population = [
    Trader(
        trader_id=i,
        sugar=random.uniform(10, 30),
        spice=random.uniform(10, 30),
        metabolism_sugar=random.randint(1, 3),
        metabolism_spice=random.randint(1, 3),
    )
    for i in range(5)
]

print("Initial population:")
for trader in population:
    print(f"  {trader}  MRS={trader.mrs:.2f}")

# Run 3 consumption steps
for step in range(3):
    for trader in population:
        if trader.alive:
            trader.consume()

print("\nAfter 3 steps of consumption:")
for trader in population:
    print(f"  {trader}")
```

Add a `harvest(sugar_amount, spice_amount)` method to the `Trader` class above that adds resources to the trader's endowment and sets `alive = True` if both endowments become positive. Test it by harvesting from a dead trader.

```python
# Copy the Trader class above and add the harvest method
```

```python
#@title Solution
class TraderV2(Trader):
    def harvest(self, sugar_amount, spice_amount):
        self.sugar += sugar_amount
        self.spice += spice_amount
        if self.sugar > 0 and self.spice > 0:
            self.alive = True

t = TraderV2(99, sugar=0.5, spice=0.5, metabolism_sugar=2, metabolism_spice=2)
t.consume()  # kills the trader
print(f"After consume: {t}")
t.harvest(sugar_amount=10, spice_amount=8)
print(f"After harvest: {t}")
```
